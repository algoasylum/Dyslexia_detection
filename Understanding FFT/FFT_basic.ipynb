```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack  import fft
from scipy.fftpack import ifft
import cmath
import plotly
import plotly.express as px
```

```python
Fs = 2
t = np.linspace(0,400,800)
f = 1/100
x =(np.cos(2*np.pi*10*t))
y =(np.sin(2*np.pi*10*t))
plt.plot(t,x)
#plt.plot(t,y)
plt.title('Sine')
```

```python
z=[]
for i in range(0,800):
    z.append(complex(x[i],-1*y[i]))
    
```

```python
Z = np.fft.fft(z)

fft_im=[]
fft_r =[]
for a in Z:
    fft_im.append(a.imag)
    fft_r.append(a.real)
```

```python
mag=[]
phase =[]
for a in Z:
    mag.append(abs(a))
    phase.append(())
```

```python
fig, ax = plt.subplots(2,2,figsize =(15,15))
ax[0,0].plot(t,fft_im)

ax[0,1].plot(t,fft_r)

ax[1,0].plot(t,mag)

#ax[1,1].plot(t,phase)
```
