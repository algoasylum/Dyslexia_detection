```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd
import math
import copy 
from matplotlib.gridspec import GridSpec

plt.style.use('seaborn')
```

## 1. Single Frequency:
### 1.1. Taking FFT for a 5Hz signal (N = 100) as the real component. 

```python
N = 100

f1 = 5
t = np.linspace(0, 1, N)
s = np.sin(f1 * 2 * np.pi * t)

fft1 = np.fft.fft(s)
```

```python
# Setting up the plot surface
fig = plt.figure(figsize=(15, 5))
gs = GridSpec(nrows=2, ncols=3)
gs.update(wspace = 0.2, hspace = 0.3)

ax0 = fig.add_subplot(gs[1, 1])
ax0.plot(range(N), fft1.imag)
ax0.set_title('Imag Component of FFT')

ax1 = fig.add_subplot(gs[0, 1])
ax1.plot(range(N), fft1.real)
ax1.set_title('Real Component of FFT')

ax0 = fig.add_subplot(gs[1, 2])
ax0.plot(range(N), np.angle(s))
ax0.set_title('Angle of input signal')

ax1 = fig.add_subplot(gs[0, 2])
ax1.plot(range(N), np.angle(fft1))
ax1.set_title('Angle of FFT')

ax2 = fig.add_subplot(gs[:, 0])
ax2.plot(range(N), s, linewidth = 0.5)
ax2.set_title('Real = '+ str(f1) + 'Hz, Imag = 0')
plt.show()
```

### 1.2. Taking FFT for a 5Hz signal (N = 100) as the imaginary component. 

```python
s_i = [complex(a,b) for a,b in zip(np.zeros(N) ,s)]
fft2 = np.fft.fft(s_i)
```

```python
# Setting up the plot surface
fig = plt.figure(figsize=(10, 5))
gs = GridSpec(nrows=2, ncols=2)
gs.update(wspace = 0.2, hspace = 0.3)

ax0 = fig.add_subplot(gs[1, 1])
ax0.plot(range(N), fft2.imag)
ax0.set_title('Imag Component of FFT')

ax1 = fig.add_subplot(gs[0, 1])
ax1.plot(range(N), fft2.real)
ax1.set_title('Real Component of FFT')

ax2 = fig.add_subplot(gs[:, 0])
ax2.plot(range(N), s, linewidth = 0.5)
ax2.set_title('Imag = '+ str(f1) + 'Hz, Real = 0')

plt.show()
```

### Conclusion: 
1. Real component is mostly positive, Imag is both positive and negative 
2. Imag component is mostly positive, Real is both positive and negative 

When the real component and the imaginary component of the input signal are interchanged: The real component and the imaginary component also interchange in the frequency spectrum 

## 2. Multiple Frequency:
### 2.1. Taking FFT for a 5Hz signal (N = 100) as the real component and 10Hz as the imaginary component. 

```python
N = 100

f1 = 5
f2 = 10

t = np.linspace(0, 1, N)
s1 = np.sin(f1 * 2 * np.pi * t) 
c1 = np.sin(f2 * 2 * np.pi * t) 
com_1 = [complex(a,b) for a,b in zip(s1 ,c1)]

fft2 = np.fft.fft(com_1)
```

```python
# Setting up the plot surface
fig = plt.figure(figsize=(15, 5))
gs = GridSpec(nrows=2, ncols=3)
gs.update(wspace = 0.2, hspace = 0.3)

ax0 = fig.add_subplot(gs[1, 1])
ax0.plot(range(N), fft2.imag)
ax0.set_title('Imag Component of FFT')

ax1 = fig.add_subplot(gs[0, 1])
ax1.plot(range(N), fft2.real)
ax1.set_title('Real Component of FFT')

ax0 = fig.add_subplot(gs[1, 2])
ax0.plot(range(N), np.angle(com_1))
ax0.set_title('Angle of input signal')

ax1 = fig.add_subplot(gs[0, 2])
ax1.plot(range(N), np.angle(fft2))
ax1.set_title('Angle of FFT')

ax2 = fig.add_subplot(gs[0, 0])
ax2.plot(range(N), np.array(com_1).real, linewidth = 0.5)
ax2.set_title('Real = ' + str(f1) + 'Hz')

ax2 = fig.add_subplot(gs[1, 0])
ax2.plot(range(N), np.array(com_1).imag, linewidth = 0.5)
ax2.set_title('Imag = ' + str(f2) + 'Hz')
plt.show()
```
