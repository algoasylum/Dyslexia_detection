```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack  import fft
from scipy.fftpack import ifft
import cmath
import plotly
import plotly.express as px
```

```python
Fs = 2
t = np.linspace(0,0.5,400)
f = 1/100
x =(np.sin(2*np.pi*10*t))+(np.sin(2*np.pi*20*t))
y =(np.sin(2*np.pi*50*t))+(np.sin(2*np.pi*60*t))
plt.plot(t,x)
#plt.plot(t,y)
plt.title('Sine')
```

```python
y1 = 20 *(np.ones(800))
y2 = x
y3 = 20*( np.sin(2*np.pi*f*t))
y4 =20*( np.sin((4*np.pi*f*t)/3))
plt.plot(y1)
plt.plot(y2)
plt.plot(y3)
plt.plot(y4)
```

```python
z1=[]
z2=[]
z3=[]
z4=[]
for i in range(0,800):
    z1.append(complex(x[i],y1[i]))
    z2.append(complex(x[i],y2[i]))
    z3.append(complex(x[i],y3[i]))
    z4.append(complex(x[i],y4[i]))
    
```

```python
N = np.size(t)
freq = (np.linspace(0,400,N))*(Fs/2)
np.size(freq)
```

```python
Z1 = np.fft.fft(z1)
Z2 = np.fft.fft(z2)
Z3 = np.fft.fft(z3)
Z4 = np.fft.fft(z4)
#ifft = np.fft.ifft(fft)
#Z1 = abs((fft[0:np.size(freq)]))
zi1 = np.fft.ifft(Z4)
zi2 = np.fft.ifft(Z4)
zi3 = np.fft.ifft(Z4)
zi4 = np.fft.ifft(Z4)
```

```python
fft1_im=[]
fft1_r =[]
for a in Z1:
    fft1_im.append(a.imag)
    fft1_r.append(a.real)
```

```python
fft2_im=[]
fft2_r =[]
for a in Z2:
    fft2_im.append(a.imag)
    fft2_r.append(a.real)
```

```python
fft3_im=[]
fft3_r =[]
for a in Z3:
    fft3_im.append(a.imag)
    fft3_r.append(a.real)
```

```python
fft4_im=[]
fft4_r =[]
for a in Z4:
    fft4_im.append(a.imag)
    fft4_r.append(a.real)
```

```python
fig, ax = plt.subplots(2,2,figsize =(15,15))
ax[0,0].plot(freq,((2/N)*abs(Z1)))
ax[0,0].set_title('Y = constant')
ax[0,1].plot(freq,((2/N)*abs(Z2)))
ax[0,1].set_title('Y = X')
ax[1,0].plot(freq,((2/N)*abs(Z3)))
ax[1,0].set_title('Y = low amplitude')
ax[1,1].plot(freq,((2/N)*abs(Z4)))
ax[1,1].set_title('Y = low amp + low freq')
```
