```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack  import fft
from scipy.fftpack import ifft

```

```python
Fs = 1
t = np.arange(0,400,1/Fs)
f = 1/100
x =50*( np.sin(2*np.pi*f*t))

plt.plot(t,x)
plt.title('Sine')
```

```python
N = np.size(t)
freq = (np.linspace(0,400,N/2))*(Fs/2)
np.size(freq)
```

```python

X = fft(x)

X1 = (2/N)*abs((X[0:np.size(freq)]))
plt.plot(freq,X1)
plt.title('fourier')
```

```python
X3  =  np.insert(X,200,np.zeros(100))
np.size(X3)
```

```python
x_reformed = ifft(X3)
```

```python
t_long = np.arange(0,500,1/Fs)
plt.plot(t_long,x_reformed)
plt.title('Reconstructed')
```
